```python
df = pd.read_csv('https://raw.githubusercontent.com/dm-fedorov/pandas_basic/master/data/it_new.csv')
df.head(3)
```

```python
s=0
k=0
for i in range(0,len(df)):
    if df.curenc[i]=='₽':
        s=s+df.col[i]
        k+=1
    if df.curenc[i]=='€': 
        s=s+df.col[i]*currencies['Valute']["EUR"]["Value"]
        k+=1
    if df.curenc[i]=='$':
        s=s+df.col[i]*currencies['Valute']["USD"]["Value"]
        k+=1
```

```python
Mean=round(s/k,0)
Mean
```
